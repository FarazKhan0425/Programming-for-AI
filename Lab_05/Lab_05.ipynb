```python
## Reading an image

import cv2

image = cv2.imread('speedy-gonzales.jpeg.scaled500.jpg')

# Extracting the height and width of an image
h, w = image.shape[:2]
# Displaying the height and width
print("Height = {}, Width = {}".format(h, w))
```

```python
## Extracting the RGB Values of a Pixel


(B, G, R) = image[100, 100]

# Displaying the pixel values
print("R = {}, G = {}, B = {}".format(R, G, B))
# We can also pass the channel to extract
B = image[100, 100, 0]
print("B = {}".format(B))
```

```python
# Extracting the Region of Interest (ROI)

from google.colab.patches import cv2_imshow

roi = image[100:500, 200:700]
cv2_imshow(roi)
```

```python
# Resizing the Image
# resize() function takes 2 parameters,
# the image and the dimensions
resize = cv2.resize(image, (500, 500))
cv2_imshow(resize)
cv2.waitKey(0)
```

```python
# Drawing a Rectangle
# Copy image to avoid modifying the original
output = image.copy()

# Ensure the top-left corner is correctly defined
new = cv2.rectangle(output, (1500, 900),(600, 400), (255, 0, 0), 2)

# Display the image
new
```

```python
# Displaying text

# Copying the original image
output = image.copy()

# Adding the text using putText() function
text = cv2.putText(output, 'OpenCV Demo', (500, 550), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 0, 0), 2)

text
```

```python
# Image resizing
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('speedy-gonzales.jpeg.scaled500.jpg')
# Loading the image

half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))

stretch_near = cv2.resize(image, (780, 540),
               interpolation = cv2.INTER_LINEAR)


Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()
```

```python
# Blurring an Image
```

```python
import numpy as np
image1 = cv2.imread('speedy-gonzales.jpeg.scaled500.jpg')

if image1 is None:
    print("Error: Image not found. Check the file path.")
else:
    print("Original Image:")
    cv2_imshow(image1)
    # Gaussian Blur
    Gaussian = cv2.GaussianBlur(image1, (7, 7), 0)
    print("Gaussian Blurring:")
    cv2_imshow(Gaussian)
    # Median Blur
    median = cv2.medianBlur(image1, 5)
    print("Median Blurring:")
    cv2_imshow(median)
    # Bilateral Blur
    bilateral = cv2.bilateralFilter(image1, 9, 75, 75)
    print("Bilateral Blurring:")
    cv2_imshow(bilateral)
```

```python
# Image Pyramid
```

```python
img = cv2.imread('speedy-gonzales.jpeg.scaled500.jpg')

# Check if the image loaded successfully
if img is None:
    print("Error: Image not found. Check the file path.")
else:
    layer = img.copy()

    plt.figure(figsize=(10, 10))

    for i in range(4):
        layer = cv2.pyrDown(layer)  # Reduce resolution

        # Convert BGR to RGB for proper Matplotlib display
        rgb_layer = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB)

        # Display with Matplotlib
        plt.subplot(2, 2, i + 1)
        plt.imshow(rgb_layer)
        plt.title(f"Layer {i + 1}")
        plt.axis("off")

        # Also display in Colab using cv2_imshow
        print(f"Displaying Layer {i + 1}:")
        cv2_imshow(layer)

    plt.show()
```
