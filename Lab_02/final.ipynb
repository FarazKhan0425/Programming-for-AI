```python
import pandas as pd #type:ignore
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head(5)
```

```python
df = df.dropna(subset=['Name'])
df.shape
```

```python
df
```

```python
df['HomePlanet'] = df['HomePlanet'].map({'Europa': 1,
                                           'Earth': 2,
                                           'Mars': 3})
df.HomePlanet[:10]
```

```python
df.Destination = df.Destination.map({'TRAPPIST-1e': 1,
                                                     '55 Cancri e': 2,
                                                     'PSO J318.5-22': 3})
df.Destination[:10]
```

```python
df['CryoSleep'] = df['CryoSleep'].fillna(df['CryoSleep'].mode()[0])
df['VIP'] = df['VIP'].fillna(df['VIP'].mode()[0])
```

```python
df['Destination'] = df['Destination'].fillna(df['Destination'].mode()[0])
df['HomePlanet'] = df['HomePlanet'].fillna(df['HomePlanet'].mode()[0])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
y = df.Transported
cols = ['PassengerId', 'HomePlanet', 'CryoSleep', 'Destination', 'Age', 'VIP', 'Name']
X = df[cols]
```

```python
from sklearn.model_selection import train_test_split #type:ignore
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor #type:ignore
from sklearn.metrics import mean_absolute_error    #type:ignore

model = RandomForestRegressor(random_state=1)
model.fit(train_X, train_y)
pred = model.predict(val_X)
val_mae = mean_absolute_error(pred, val_y)

print("Validation: {:,.0f}".format(val_mae))
```
